# Error reading from pipe

We redirect stderr to /dev/null to avoid difference between ipykernel 5 (that sends stderr to the console) and ipykernel 6 (that captures stderr for the notebook).

```python
spot.automaton('non-existing-cmd 2>/dev/null |')
```

```python
spot.automaton('sleep 3; cat _example.aut |', timeout=1)
```

```python
for a in spot.automata("ltl2tgba 'a U b'|", 'ltl2tgba "syntax U U error" 2>/dev/null |'):
    display(a)
```

Reading an empty file with `spot.automaton()` is an error.

```python
spot.automaton('true|')
```
