```python
import spot
from spot.jupyter import display_inline
spot.setup()

def test_and(left, right):
    res = spot.product_susp(left, right)
    display_inline(left, right, res)
    assert res.equivalent_to(spot.product(left, right))

def test_or(left, right):
    res = spot.product_or_susp(left, right)
    display_inline(left, right, res)
    assert res.equivalent_to(spot.product_or(left, right))

def test(left, right):
    test_and(left, right)
    test_or(left, right)
```

```python
left = spot.translate('(Ga | Gb | Gc)')
right = spot.translate('GFa')
test(left, right)
```

```python
left = spot.translate('(Ga | Gb | Gc)', 'det')
right = spot.translate('GFa')
test(left, right)
```

```python
test(spot.translate('Ga | G!a', 'det'), spot.translate('GF(a <-> !Xa)'))
```

```python
a = spot.translate('(Ga | GF!a)', 'det', 'gen')
test(a, spot.translate('GF(a <-> !Xa)'))
```

```python
a = spot.translate('(Ga | GF!a)')
tt = spot.translate('true', 'monitor')
test(a, tt)
```

```python
tt.set_acceptance(spot.acc_cond(1, "t"))  # issue 546
test(a, tt)
```

```python
tt.set_acceptance(spot.acc_cond(1, "Fin(1) | Inf(1)"))  # issue 546
test(a, tt)
```

```python
tt.set_acceptance(spot.acc_cond(1, "Fin(1) & Inf(1)"))
test(a, tt)
```

```python
test(a, spot.translate('false'))
```
